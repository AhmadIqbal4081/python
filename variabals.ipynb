```python
class car():
    def __init__(self, make, model, color, year, ownerInfo):
        self.make = make
        self.model = model
        self.color = color
        self.year = year
        self.fuel = 0
        self.mileage = 0
        self.ownerInfo = ownerInfo

    def fuel_up(self, liters):
        self.fuel += liters
        print(f"Fueled up {liters} liters. Total fuel: {self.fuel} liters.")

    def getcurrent_fuel(self):
        print(f"Current fuel level: {self.fuel} liters.")
        return self.fuel

    def get_owner_info(self):
        return self.ownerInfo

    def getCurrent_mileage(self):
        print(f"Current mileage: {self.mileage} km.")
        return self.mileage

    def drive(self, distance):
        if self.fuel <= distance :
            print("Not enough fuel to drive.")
            return
        self.mileage += distance
        self.fuel -= distance / 10  # 10 km per liter
        print(f"Drove {distance} km. Remaining fuel: {self.fuel} liters.")


car_details = car("Toyota", "Camry", "Black", 2020, {"name": "Ahmad"})

car_details.fuel_up(50)
car_details.getcurrent_fuel()
car_details.drive(100)
car_details.getCurrent_mileage()
car_details.drive(250)
car_details.getCurrent_mileage()

car_details.get_owner_info()
```

```python
class Bike():
    def __init__(self, make, model, color, year, ownerInfo):
        self.make = make
        self.model = model
        self.color = color
        self.year = year
        self.fuel = 0
        self.mileage = 0
        self.ownerInfo = ownerInfo

    def fuel_up(self, liters):
        self.fuel += liters
        print(f"Bike fueled up with {liters} liters. Total fuel: {self.fuel} liters.")

    def getcurrent_fuel(self):
        print(f"Current bike fuel level: {self.fuel} liters.")
        return self.fuel

    def get_owner_info(self):
        return self.ownerInfo

    def getCurrent_mileage(self):
        print(f"Current bike mileage: {self.mileage} km.")
        return self.mileage

    def ride(self, distance):
        if self.fuel <= distance / 40:  # bike fuel average 40 km per liter
            print("Not enough fuel to ride.")
            return
        self.mileage += distance
        self.fuel -= distance / 40
        print(f"Rode {distance} km. Remaining fuel: {self.fuel} liters.")
```

```python
bike1 = Bike("Honda", "125", "Red", 2023, {"name": "Ali"})

bike1.fuel_up(5)
bike1.getcurrent_fuel()
bike1.ride(60)
bike1.getCurrent_mileage()
bike1.ride(120)
bike1.get_owner_info()
```

```python
class Laptop():
    def __init__(self, brand, model, color, year, ownerInfo):
        self.brand = brand
        self.model = model
        self.color = color
        self.year = year
        self.battery = 100   # battery percentage
        self.usage_hours = 0
        self.ownerInfo = ownerInfo

    def charge(self, percent):
        self.battery += percent
        if self.battery > 100:
            self.battery = 100
        print(f"Laptop charged. Current battery: {self.battery}%")

    def get_battery_level(self):
        print(f"Battery level: {self.battery}%")
        return self.battery

    def get_owner_info(self):
        return self.ownerInfo

    def get_usage_hours(self):
        print(f"Used for: {self.usage_hours} hours.")
        return self.usage_hours

    def use(self, hours):
        if self.battery <= hours * 10:  # 10% battery per hour
            print("Not enough battery to use laptop for that long.")
            return
        self.usage_hours += hours
        self.battery -= hours * 10
        print(f"Used laptop for {hours} hours. Remaining battery: {self.battery}%")
```

```python
laptop1 = Laptop("Dell", "Inspiron 15", "Silver", 2021, {"name": "Ahmed"})

laptop1.get_battery_level()
laptop1.use(3)
laptop1.get_usage_hours()
laptop1.charge(20)
laptop1.get_owner_info()
```
